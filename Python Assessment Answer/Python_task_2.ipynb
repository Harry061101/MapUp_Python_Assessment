```python
import pandas as pd

def calculate_distance_matrix(df):
    distance_matrix = pd.DataFrame()
    unique_ids = sorted(set(df['id_start']).union(set(df['id_end'])))
    
    for id_start in unique_ids:
        start_rows = df[df['id_start'] == id_start]

        for id_end in unique_ids:
            end_rows = df[df['id_end'] == id_end]
            common_links = pd.merge(start_rows, end_rows, left_on='id_end', right_on='id_start', suffixes=('_start', '_end'))
            cumulative_distance = common_links['distance_start'] + common_links['distance_end']
            min_distance = cumulative_distance.min()
            distance_matrix.at[id_start, id_end] = min_distance
    distance_matrix = distance_matrix.fillna(0)
    distance_matrix = distance_matrix + distance_matrix.transpose()

    return distance_matrix

result_matrix = calculate_distance_matrix(df)
print(result_matrix)
```

```python
import pandas as pd
df=pd.read_csv('dataset-3.csv')
def unroll_distance_matrix(df):
    unrolled_df = pd.DataFrame(columns=['id_start', 'id_end', 'distance'])
    for id_start in df.index:
        for id_end in df.columns:
            if id_start != id_end:
                distance = df.at[id_start, id_end]
                unrolled_df = unrolled_df.append({'id_start': id_start, 'id_end': id_end, 'distance': distance}, ignore_index=True)

    return unrolled_df
distance_matrix = calculate_distance_matrix(df) 
unrolled_result = unroll_distance_matrix(distance_matrix)
print(unrolled_result)
```
